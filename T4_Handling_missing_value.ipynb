```python
# prompt: read file from drive using pyspark

import pyspark
from google.colab import drive

drive.mount('/content/drive')

# spark = pyspark.SparkContext()
df = spark.read.csv('/content/drive/My Drive/Colab Notebooks/data_t4.csv', header=True)
df.show()
```

```python
#drop the column

df.drop('name').show()
```

```python
#drop rows which includes null
# .drop(how='all') means if all the values in row are null then only delete row
# .drop(how='any') means if there is even one null is present then also delete row > this is by default parameter
# .drop(how='all/any', thresh=2) means if there is less than 2 non null value then only row deleted
df.na.drop().show()
```

```python
#only specify column has null value then deleted others are not
df.na.drop(subset=['age']).show()
```

```python
df.show()
```

```python
### fill the missing value

df.na.fill('Missing val').show()
```

```python
df.na.fill('Missing val','experience').show()
```

```python
df.na.fill('Missing val',['experience','age']).show()
```

```python
df.show()
```

```python
# fill value using mean of that column

#so for that we need to make imputer (kind of function)

from pyspark.ml.feature import Imputer
imputer= Imputer(
    inputCols=['age','experience','salary'],
    outputCols=["{}_imputed".format(c) for c in ['age','experience','salary']]
).setStrategy("mean")
```

```python
# prompt: change age experience and salary column from string to integer in data using infer

df = spark.read.csv('/content/drive/My Drive/Colab Notebooks/data_t4.csv', header=True, inferSchema=True)
df.show()
```

```python
imputer.fit(df).transform(df).show()
```
